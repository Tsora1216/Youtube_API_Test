```python
f = open('../../Key/Youtube_API_Key.txt', 'r')
API_KEY = f.read()
f.close()
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_VERSION = 'v3'
SEARCH_TEXT = 'ユースフル / スキルの図書館'

youtube = build(
    YOUTUBE_API_SERVICE_NAME,
    YOUTUBE_API_VERSION,
    developerKey=API_KEY
)

response = youtube.search().list(q=SEARCH_TEXT, part='id,snippet', maxResults=25).execute()

for item in response.get('items', []):
    if item['id']['kind'] != 'youtube#channel':
        continue
    print('*' * 10)
    CHANNEL_ID = item['snippet']['channelId']
    print(CHANNEL_ID)
    json_data =json.dumps(item, indent=2, ensure_ascii=False)
    print(json.dumps(item, indent=2, ensure_ascii=False))
    print(json_data)
    print('*' * 10)
```

```python
import pandas as pd
df = pd.DataFrame(columns=["id", "title", "viewCount", "likeCount", "favoriteCount", "commentCount", "publishedAt"])
df
```

```python
from datetime import datetime

for video_id in VIDEO_ID_LIST:
    response = youtube.videos().list(
      part = 'snippet,statistics',
      id = video_id
      ).execute()

    for item in response.get("items", []):
        if item["kind"] != "youtube#video":
            continue
        #print('*' * 10)
        #print(json.dumps(item, indent=2, ensure_ascii=False))
        item_id=item["id"]
        item_title=item["snippet"]["title"]
        item_viewCount=item["statistics"]["viewCount"]
        item_likeCount=item["statistics"]["likeCount"]
        item_favoriteCount=item["statistics"]["favoriteCount"]
        item_commentCount=item["statistics"]["commentCount"]
        date_str = item["snippet"]["publishedAt"]
        date_time = datetime.fromisoformat(date_str)
        item_publishedAt=date_time.strftime("%Y-%m-%d %H:%M:%S")
        new_data = pd.DataFrame({'id': item_id, 'title': item_title, 'viewCount': item_viewCount,'likeCount': item_likeCount,'favoriteCount': item_favoriteCount,'commentCount': item_commentCount,'publishedAt': item_publishedAt}, index=[0])
        df = pd.concat([df, new_data], ignore_index=True)
        #print('*' * 10)
```

```python
# ライブラリのインポート
import sqlite3

# データベース名とテーブル名
db_name = 'youtube4.db'
table_name = 'data'

# SQLiteに書き込む
with sqlite3.connect(db_name) as conn:
    df.to_sql(table_name, conn)
```
